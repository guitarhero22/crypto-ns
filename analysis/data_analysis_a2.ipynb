```python
import pandas as pd
import os.path
import seaborn
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = np.linspace(0.2, 0.5,100)
gammas = [0.25, 0.5, 0.75]
results_mpu_theory = {}
for gamma in gammas:
    numerator = alpha * ((1-alpha)**2)*(4*alpha + gamma*(1-2*alpha)) - (alpha**3)
    denominator =1 - (alpha*(1 + (2-alpha)*alpha))
    results_mpu_theory[int(gamma*100)] = numerator/denominator
    
```
